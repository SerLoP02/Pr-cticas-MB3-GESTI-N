Google ADK define dos tipos de instructions (global_instruction está deprecated)

# Instructions

Cuando solo se provee del parámetro `instructions`, este actuará como `static_instruction`. 

# Static Instructions

`static_instruction` está diseñado para ser inmutable. Es el primer mensaje que se envía al modelo y donde se definen reglas que NO van a cambiar. Se ejecuta al inicio de una sesión.
