Se define un Run de un agente como el ciclo:
1. Usuario hace una pregunta
2. LLM decide si llamar a una herramienta; si no, el run se acaba aquí
3. Si se llama a una herramienta, se ejecuta dicha herramienta
4. El output de la herramienta se manda a la LLM y decide qué hacer
5. LLM decide si llamar a una herramienta; si no, el run se acaba aquí
6. ...

Una vez definido el ciclo o run del agente, podemos pasar a describir los numerosos callbacks que nos ofrece Google ADK

* **Before/After Agent Callback**: Se ejecuta SIEMPRE antes/después de cada ciclo del agente
* **Before/After Model Callback**: Se ejecuta SIEMPRE antes/después de cada paso por la LLM
* **Before/After Tool Callback**: Se ejecuta SIEMPRE antes/después de cada ejecución de la herramienta que la LLM ha llamado

Para más información y ejemplos, revisar la documentación adjuntada
